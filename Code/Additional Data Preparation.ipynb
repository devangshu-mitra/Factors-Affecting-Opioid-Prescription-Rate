```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder

# Define the classification groups
classification_groups = {
    'Primary Care': ['Family Practice', 'Internal Medicine', 'Pediatric Medicine', 'General Practice', 'Family Medicine'],
    'Specialty Care': ['Obstetrics & Gynecology', 'Optometry' , 'Clinical Neuropsychologist', 'Psychiatry', 'Neurology', 'Psychiatry & Neurology', 'Medical Genetics and Genomics', 'Physical Medicine and Rehabilitation', 'Nephrology', 'Infectious Disease', 'Anesthesiology', 'Interventional Pain Management', 'Sleep Medicine', 'Urology', 'Maxillofacial Surgery', 'Rheumatology', 'Osteopathic Manipulative Medicine', 'Geriatric Medicine', 'Cardiology', 'Thoracic Surgery', 'Pulmonary Disease', 'Gastroenterology', 'Medical Oncology', 'Vascular Surgery', 'Hospice and Palliative Care', 'Plastic Surgery', 'Allergy/ Immunology', 'Sports Medicine', 'Podiatry', 'Plastic and Reconstructive Surgery', 'Ophthalmology', 'Clinical Cardiac Electrophysiology', 'Dermatology', 'Pain Medicine', 'Hand Surgery', 'Endocrinology', 'Radiation Oncology', 'Interventional Cardiology', 'Colorectal Surgery (Proctology)', 'Cardiac Surgery', 'Surgical Oncology', 'Neurosurgery', 'Gynecological Oncology', 'Thoracic Surgery (Cardiothoracic Vascular Surgery)', 'Interventional Radiology', 'Micrographic Dermatologic Surgery', 'Neurological Surgery', 'Geriatric Psychiatry', 'Legal Medicine', 'Undersea and Hyperbaric Medicine', 'Hematopoietic Cell Transplantation and Cellular Therapy', 'Advanced Heart Failure and Transplant Cardiology', 'Psychoanalyst', 'Independent Medical Examiner'],
    'Nursing': ['Nurse Practitioner', 'Certified Clinical Nurse Specialist', 'Certified Nurse Midwife', 'Registered Nurse', 'Certified Registered Nurse Anesthetist (CRNA)', 'Licensed Practical Nurse'],
    'Other Healthcare': ['Physician Assistant', 'Behavior Analyst',  'Marriage & Family Therapist', 'Pharmacist', 'Student in an Organized Health Care Education/Training Program', 'Diagnostic Radiology', 'Peripheral Vascular Disease', 'Nuclear Medicine', 'Physical Therapist in Private Practice', 'Neuromusculoskeletal Medicine, Sports Medicine', 'Chiropractic', 'Dental Assistant', 'Counselor', 'Naturopath', 'Colon & Rectal Surgery', 'Psychologist, Clinical', 'Psychologist', 'Emergency Medical Technician, Basic'],
    'Unknown': ['Undefined Physician type', 'Unknown Supplier/Provider Specialty'],
    'Clinic/Center': ['Clinic/Center', 'Clinic or Group Practice', 'Health Maintenance Organization'],
    'Emergency Medicine': ['Emergency Medicine', 'Ambulatory Surgical Center'],
    'Hospitalist': ['Hospitalist'],
    'Other': ['Specialist','Social Worker']
}

# Create a function to assign each specialty to a classification group
def assign_classification(specialty):
    for group, specialties in classification_groups.items():
        if specialty in specialties:
            return group
    return 'Unknown'

# Apply the function to the 'Prscrbr_Type' column to create a new column 'Classification'
data['Classification'] = data['Prscrbr_Type'].apply(assign_classification)

```

```python
# Use LabelEncoder to convert the 'Classification' column to numeric values
le = LabelEncoder()
data['Classification_Encoded'] = le.fit_transform(data['Classification'])

# Print the first 5 rows of the data with the new columns
print(data.head())
```

```python
# Save the filtered dataset to a new CSV file
data.to_csv('calc_new_dataset.csv', index=False)
```
